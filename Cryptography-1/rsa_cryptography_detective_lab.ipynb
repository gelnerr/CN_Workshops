## 1. Why Do We Need Cryptography?
When you send a message online, it can pass through many computers.

- We want our messages to stay **secret**.
- We want to know they **weren't changed**.
- We want to know **who** really sent them.

Cryptography is like giving your message a **lock** so only the right person can open it.

## 2. The Big Idea of RSA (No Heavy Math Yet)
RSA uses **two keys**:

- A **public key**: like a padlock you can give to everyone.
- A **private key**: like the only key that opens that padlock.

Anyone can use your public key to lock (encrypt) a message.
But only you, with your private key, can unlock (decrypt) it.

In this notebook, we'll build a tiny version of RSA so you can see how it works.

## 3. Step 1 â€“ Choose Two Prime Numbers
Real RSA uses **huge** prime numbers. For learning, we'll use small ones so you can follow the steps.

We'll pick:

- `p = 61`
- `q = 53`

From these, we compute:

- `n = p Ã— q` (part of both keys)
- `phi = (p - 1) Ã— (q - 1)` (used to build the keys)

```python
p = 61  # first prime
q = 53  # second prime

n = p * q
phi = (p - 1) * (q - 1)

print("p =", p)
print("q =", q)
print("n = p * q =", n)
print("phi = (p - 1) * (q - 1) =", phi)
```

## 4. Step 2 â€“ Choose the Public Exponent `e`
The number `e` is part of the **public key**. It must:

- Be smaller than `phi`.
- Not share any common factors with `phi` (except 1).

A common and simple choice is `e = 17`, which works well with our `phi`.

```python
e = 17  # public exponent
print("e =", e)
```

## 5. Step 3 â€“ Find the Private Exponent `d`
The number `d` is part of the **private key**.

It is chosen so that:

`(d Ã— e) % phi == 1`

In other words, `d` is the **modular inverse** of `e` with respect to `phi`.
We'll search for it with a simple loop (good enough for our small example).

```python
def mod_inverse(e, phi):
    """Find d such that (d * e) % phi == 1."""
    for d in range(1, phi):
        if (d * e) % phi == 1:
            return d
    return None

d = mod_inverse(e, phi)
print("d =", d)
print("Public key  = (e, n) = (", e, ",", n, ")")
print("Private key = (d, n) = (", d, ",", n, ")")
```

## 6. Step 4 â€“ Turn Text into Numbers
RSA works with **numbers**, not letters.

We'll use a simple code:

- `A â†’ 1`, `B â†’ 2`, ..., `Z â†’ 26`

We'll also remove spaces and make everything uppercase to keep it simple.

```python
def text_to_numbers(text: str):
    """Convert Aâ€“Z to 1â€“26 and remove spaces."""
    text = text.upper().replace(" ", "")
    nums = []
    for ch in text:
        if 'A' <= ch <= 'Z':
            nums.append(ord(ch) - 64)
    return nums

example = "HELLO"
num_example = text_to_numbers(example)
print(example, "â†’", num_example)
```

## 7. Step 5 â€“ Encrypt Numbers with the Public Key
To **encrypt** a number `m` (a piece of the message), RSA does:

`cipher = (m ** e) % n`

We'll write a function that encrypts a whole list of numbers.

```python
def rsa_encrypt(numbers, e, n):
    """Encrypt a list of numbers using RSA."""
    cipher_nums = []
    for m in numbers:
        c = (m ** e) % n
        cipher_nums.append(c)
    return cipher_nums

cipher_example = rsa_encrypt(num_example, e, n)
print("Encrypted HELLO â†’", cipher_example)
```

## 8. Step 6 â€“ Decrypt Numbers with the Private Key
To **decrypt** a number `c` (a piece of the encrypted message), RSA does:

`m = (c ** d) % n`

We'll write a function that decrypts a whole list of numbers.

```python
def rsa_decrypt(cipher_nums, d, n):
    """Decrypt a list of RSA-encrypted numbers."""
    plain_nums = []
    for c in cipher_nums:
        m = (c ** d) % n
        plain_nums.append(m)
    return plain_nums

decrypted_example = rsa_decrypt(cipher_example, d, n)
print("Decrypted back â†’", decrypted_example)
```

## 9. Step 7 â€“ Turn Numbers Back into Text
Now we turn our 1â€“26 numbers back into letters:

- `1 â†’ A`, `2 â†’ B`, ..., `26 â†’ Z`

```python
def numbers_to_text(nums):
    """Convert 1â€“26 numbers back to Aâ€“Z text."""
    chars = []
    for n_val in nums:
        if 1 <= n_val <= 26:
            chars.append(chr(n_val + 64))
        else:
            chars.append('?')  # out of range
    return ''.join(chars)

print("Numbers â†’ text:", numbers_to_text(decrypted_example))
```
